```python
df['Dates'] = pd.to_datetime(df['time']).dt.date
df['Time'] = pd.to_datetime(df['time']).dt.time
```

```python
df.drop(['time'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Dates_encoded']=pd.DataFrame(le.fit_transform(df.Dates))
df['Time_encoded']=pd.DataFrame(le.fit_transform(df.Time))
df['territoryName_encoded']=pd.DataFrame(le.fit_transform(df.territoryName))
```

```python
df.drop(['Time','territoryName','Dates'],axis=1,inplace=True)
```

```python
party_mean_encoded=pd.Series(df['Party'].value_counts())/len(df['Party'])
x=party_mean_encoded.to_dict()
df['Party_encoded']=df['Party'].map(x)
df.head()
```

```python
df.drop(['Party'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df.drop('FinalMandates',axis=1))
scaled_features=scaler.transform(df.drop('FinalMandates',axis=1))
dataset_scaled=pd.DataFrame(scaled_features,columns=df.columns[:-1])
dataset_scaled.head()
```

```python
X=df.drop(['FinalMandates'],axis=1) 
y=df['FinalMandates']



```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(model.predict(X_test),y_test)
score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
parameters={'n_neighbors':(1,10)}
gd=GridSearchCV(kn,parameters,cv=10)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
dt=DecisionTreeRegressor()
parameters={'criterion':('mse','mae','friedman_mse')}
gd=GridSearchCV(dt,parameters)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.linear_model import Lasso
```

```python
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(X,y)
clf.best_params_
```

```python
model=[]
score=[]
cvs=[]
for i in [LinearRegression(),KNeighborsRegressor(n_neighbors=10),DecisionTreeRegressor(criterion='mae'),Lasso(alpha=0.001)]:
    k=i
    model.append(i)
    print("\n")
    print("model calculation for ",i,"are:")
    k.fit(X_train,y_train)
    k.score(X_train,y_train)
    pred=k.predict(X_test)
    print(pred)
    rscore=r2_score(k.predict(X_test),y_test)
    print("r2 score is",rscore)
    score.append(rscore)
    cv_score=cross_val_score(k,X,y,cv=5).mean()
    print('cv score is',cv_score)
    cvs.append(cv_score)
```
