```python
model=[]
score=[]
cvs=[]


for i in [LogisticRegression(),KNeighborsClassifier(n_neighbors=10),DecisionTreeClassifier(criterion='entropy'),SVC(C=10,kernel='rbf')]:
    k=i
    model.append(i)
    print("\n")
    print("model calculation for ",i,"are:")
    k.fit(a_train,b_train)
    k.score(a_train,b_train)
    pred=k.predict(a_test)
    print(pred)
    AS=accuracy_score(pred,b_test)
    print("accuracy score is",AS)
    score.append(AS)
    cv_score=cross_val_score(k,a,b,cv=5,scoring='accuracy').mean()
    print('cv score is',cv_score)
    cvs.append(cv_score)
    print("")
    cm=confusion_matrix(pred,b_test)
    print(cm)
    print("")
    print(classification_report(pred,b_test))
    print("\n")
```
