```python
auto_df.drop(["Customer","Country","State Code","State","Effective To Date","Gender","Marital Status"],axis=1,inplace=True)
```

```python
auto_df.dtypes
```

```python
auto_df['Response']=auto_df['Response'].map({'No': 0, 'Yes': 1})
```

```python
auto_df['Education']=auto_df['Education'].map({'Bachelor': 0, 'College': 1,'High School or Below':2,'Master':3,'Doctor':4})
```

```python
auto_df["Vehicle Class"].value_counts()
```

```python
auto_df['EmploymentStatus']=auto_df['EmploymentStatus'].map({'Employed': 0, 'Unemployed': 1,'Medical Leave':2,'Disabled':3,'Retired':4})
```

```python
auto_df['Location Code']=auto_df['Location Code'].map({'Suburban': 0, 'Rural': 1,'Urban':2})
```

```python
auto_df.dtypes
```

```python
auto_df['Policy Type']=auto_df['Policy Type'].map({'Personal Auto': 0, 'Corporate Auto': 1,'Special Auto':2})
```

```python
auto_df['Claim Reason']=auto_df['Claim Reason'].map({'Collision': 0, 'Hail': 1,'Scratch/Dent':2,'Other':3})
```

```python
auto_df.dtypes
```

```python
auto_df.isnull().sum()
```

```python
auto_df=pd.get_dummies(auto_df,columns=['Policy','Sales Channel','Vehicle Class','Vehicle Size','Coverage'])
```

```python
auto_df.dtypes
```

```python
auto_df[["Claim Amount", "Total Claim Amount"]] = auto_df[["Claim Amount", "Total Claim Amount"]].astype(int)
```

```python
X=auto_df.drop('Claim Amount',axis=1)
y=auto_df["Claim Amount"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(model.predict(X_test),y_test)
score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
parameters={'n_neighbors':(1,10)}
gd=GridSearchCV(kn,parameters,cv=10)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
dt=DecisionTreeRegressor()
parameters={'criterion':('mse','mae','friedman_mse')}
gd=GridSearchCV(dt,parameters)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.linear_model import Lasso
```

```python
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(X,y)
clf.best_params_
```

```python
model=[]
score=[]
cvs=[]
for i in [LinearRegression(),KNeighborsRegressor(n_neighbors=10),DecisionTreeRegressor(criterion='mse'),Lasso(alpha=1)]:
    k=i
    model.append(i)
    print("\n")
    print("model calculation for ",i,"are:")
    k.fit(X_train,y_train)
    k.score(X_train,y_train)
    pred=k.predict(X_test)
    print(pred)
    rscore=r2_score(k.predict(X_test),y_test)
    print("r2 score is",rscore)
    score.append(rscore)
    cv_score=cross_val_score(k,X,y,cv=5).mean()
    print('cv score is',cv_score)
    cvs.append(cv_score)
```
