```python
continuous_vars = ['Distance from Residence to Work', 'Service time', 'Age', 'Transportation expense',
       'Hit target', 'Weight', 'Height', 'Body mass index', 'Absenteeism time in hours']

categorical_vars = ['ID','Reason for absence','Month of absence','Day of the week',
                     'Seasons','Disciplinary failure', 'Education', 'Social drinker',
                     'Social smoker', 'Pet', 'Son']
```

```python
for i in continuous_vars:
    if i == 'Absenteeism time in hours':
        continue
    df[i] = (df[i] - df[i].min())/(df[i].max()-df[i].min())
```

```python
df = pd.get_dummies(data = df, columns = categorical_vars)
```

```python
df1 = df.copy()
```

```python
df1.drop(columns=['Work load Average/day '],axis=1,inplace=True)
```

```python
df.shape
```

```python
#PCA
```

```python
target = df['Absenteeism time in hours']
```

```python
df.drop(['Absenteeism time in hours'], inplace = True, axis=1)
df.shape
```

```python
from sklearn.decomposition import PCA


X = df1.values
pca = PCA(n_components=110)
pca.fit(X)
var= pca.explained_variance_ratio_
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)

plt.plot(var1)
plt.show()
```

```python
pca = PCA(n_components=45)
pca.fit(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,target, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
pred
```

```python
from sklearn.metrics import r2_score
score=r2_score(model.predict(X_test),y_test)
score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
parameters={'n_neighbors':(1,10)}
gd=GridSearchCV(kn,parameters,cv=10)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
dt=DecisionTreeRegressor()
parameters={'criterion':('mse','mae','friedman_mse')}
gd=GridSearchCV(dt,parameters)
gd.fit(X_train,y_train)
gd.best_params_
```

```python
from sklearn.linear_model import Lasso
```

```python
lsreg=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
clf=GridSearchCV(lsreg,parameters,cv=10)
clf.fit(X,target)
clf.best_params_
```

```python
model=[]
score=[]
cvs=[]
for i in [LinearRegression(),KNeighborsRegressor(n_neighbors=10),DecisionTreeRegressor(criterion='mae'),Lasso(alpha=0.001)]:
    k=i
    model.append(i)
    print("\n")
    print("model calculation for ",i,"are:")
    k.fit(X_train,y_train)
    k.score(X_train,y_train)
    pred=k.predict(X_test)
    print(pred)
    rscore=r2_score(k.predict(X_test),y_test)
    print("r2 score is",rscore)
    score.append(rscore)
    cv_score=cross_val_score(k,X,target,cv=5).mean()
    print('cv score is',cv_score)
    cvs.append(cv_score)
```
