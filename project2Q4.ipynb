```python
#Write a python program to scrap book name, author name, genre and book review of any 5 books from
#‘www.bookpage.com’
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
book_name=[]
author_name=[]
book_genre=[]
book_review=[]
```

```python
titles_column = soup.find_all('h4', class_="italic")
```

```python
titles_column
```

```python
for information in titles_column:
        title = information.a.text
        book_name.append(title)
```

```python
book_name
```

```python
author_column = soup.find_all('p', class_="sans bold")
```

```python
author_column
```

```python
for information in author_column:
        author = information.text.strip()
        author_name.append(author)
```

```python
author_name
```

```python
genre_column = soup.find_all('p', class_='genre-links hidden-phone')
```

```python
genre_column
```

```python
for information in genre_column:
        genre = information.a.text
        book_genre.append(genre)
```

```python
book_genre
```

```python
review_column = soup.find_all('p', class_='excerpt')
```

```python
review_column
```

```python
for information in review_column:
        review = information.text.strip()
        book_review.append(review)
```

```python
book_review
```

```python
for i in range(0,5):
    print(book_name[i] ,' ,',author_name[i],',', book_genre[i],',',book_review[i])
    print('\n')
```

```python
import pandas as pd
```

```python
books=pd.DataFrame({})
books['name']=book_name[:5]
books['author']=author_name[:5]
books['genre']=book_genre[:5]
books['reviews']= book_review[:5]
```

```python
books
```

```python
books.to_csv('Books.csv')
```
